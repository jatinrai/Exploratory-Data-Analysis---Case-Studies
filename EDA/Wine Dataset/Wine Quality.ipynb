## Visualization Analysis
- Based on histograms of columns in this dataset, which of the following feature variables appear skewed to the right? Fixed Acidity, Total Sulfur Dioxide, pH, Alcohol
- Based on scatterplots of quality against different feature variables, which of the following is most likely to have a positive impact on quality? Volatile Acidity, Residual Sugar, pH, Alcohol

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2)
fig.set_size_inches(10,10)
sns.distplot(df['fixed acidity'], kde=False, ax=axs[0,0]);
sns.distplot(df['total sulfur dioxide'], kde=False, ax=axs[0,1]);
sns.distplot(df['pH'], kde=False, ax=axs[1,0]);
sns.distplot(df['alcohol'], kde=False, ax=axs[1,1]);
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(10,10)
sns.scatterplot('quality','volatile acidity',data=df,ax=axs[0,0]);
sns.scatterplot('quality','residual sugar',data=df,ax=axs[0,1]);
sns.scatterplot('quality','pH',data=df,ax=axs[1,0]);
sns.scatterplot('quality','alcohol',data=df,ax=axs[1,1]);
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2)
fig.set_size_inches(10,10)
sns.barplot('quality','volatile acidity',data=df,ax=axs[0,0]);
sns.barplot('quality','residual sugar',data=df,ax=axs[0,1]);
sns.barplot('quality','pH',data=df,ax=axs[1,0]);
sns.barplot('quality','alcohol',data=df,ax=axs[1,1]);
```

```python
df.groupby(['quality','color']).mean()
```

Q1: Is a certain type of wine (red or white) associated with higher quality?

```python
df.groupby(['color'])['quality'].mean()
```

Q2: What level of acidity (pH value) receives the highest average rating?

```python
np.percentile(df['pH'],[0,25,50,75,100])
```

```python
df['acidity_levels'] = pd.cut(df['pH'],bins=[2.72, 3.11, 3.21, 3.32, 4.01],labels=['High','Moderately High','Medium','Low'],include_lowest=True)
```

```python
df.groupby('acidity_levels')['quality'].mean()
```

Q3: Do wines with higher alcoholic content receive better ratings?

```python
np.percentile(df['alcohol'],[0,50,100])
```

```python
df['alcohol_levels'] = pd.cut(df['alcohol'],bins=[ 8. , 10.3, 14.9],labels=['Low','High'],include_lowest=True)
```

```python
df.groupby('alcohol_levels')['quality'].mean()
```

Q4: Do sweeter wines (more residual sugar) receive better ratings?

```python
np.percentile(df['residual sugar'],[0,50,100])
```

```python
df['sugar_levels'] = pd.cut(df['residual sugar'],bins=[ 0.6,  3. , 65.8],labels=['Less','More'],include_lowest=True)
```

```python
df.groupby('sugar_levels')['quality'].mean()
```

```python
color_means = df.groupby(['color'])['quality'].mean()
color_means.plot(kind='bar', title = 'Average Wine Quality By Color', colors = ['red', 'grey'], alpha = 0.7)
plt.xlabel('Colors',fontsize=13)
plt.ylabel('Quality',fontsize=13)
plt.title('Average Wine Quality By Color',fontsize=18);
```

```python
counts = df.groupby(['quality', 'color'])['pH'].count()
counts.plot(kind='bar', color = ['red','grey'], alpha=0.7);
plt.xlabel('Quality & Color', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.title('Counts by Wine Quality and Color',fontsize=18);
```

```python
totals = df.groupby('color')['pH'].count()
proportions = counts/totals
proportions.plot(kind='bar', color = ['red','grey'], alpha=0.7);
plt.xlabel('Quality & Color', fontsize=13)
plt.ylabel('Count', fontsize=13)
plt.title('Proportion by Wine Quality and Color',fontsize=18);
```
